```python
# PyTorch iÃ§in Custom Dataset sÄ±nÄ±fÄ±
class ImageDataset(Dataset):
    def __init__(self, data_dir, transform=None, train_ratio=0.8, split='train'):
        self.data_dir = Path(data_dir)
        self.transform = transform
        self.images = []
        self.labels = []
        
        # SÄ±nÄ±flarÄ± bul - sadece gÃ¶rÃ¼ntÃ¼ iÃ§eren klasÃ¶rler
        all_dirs = [d for d in self.data_dir.iterdir() if d.is_dir()]
        self.classes = []
        
        # Ã–nce direkt klasÃ¶rlerde ara
        for d in all_dirs:
            images = list(d.glob('*.jpg')) + list(d.glob('*.jpeg')) + \
                    list(d.glob('*.png')) + list(d.glob('*.JPG')) + \
                    list(d.glob('*.JPEG')) + list(d.glob('*.PNG'))
            if len(images) > 0:
                self.classes.append(d.name)
        
        # EÄŸer bulunamadÄ±ysa, bir seviye daha derine in (raw-img gibi ara klasÃ¶rler iÃ§in)
        if len(self.classes) == 0:
            for subdir in all_dirs:
                for sub_subdir in subdir.iterdir():
                    if sub_subdir.is_dir():
                        images = list(sub_subdir.glob('*.jpg')) + list(sub_subdir.glob('*.jpeg')) + \
                                list(sub_subdir.glob('*.png')) + list(sub_subdir.glob('*.JPG')) + \
                                list(sub_subdir.glob('*.JPEG')) + list(sub_subdir.glob('*.PNG'))
                        if len(images) > 0:
                            if sub_subdir.name not in self.classes:
                                self.classes.append(sub_subdir.name)
                if len(self.classes) > 0:
                    # data_dir'yi gÃ¼ncelle
                    self.data_dir = subdir
                    break
        
        self.classes = sorted(list(set(self.classes)))
        self.class_to_idx = {cls: idx for idx, cls in enumerate(self.classes)}
        
        if len(self.classes) == 0:
            raise ValueError(f"HiÃ§ gÃ¶rÃ¼ntÃ¼ iÃ§eren sÄ±nÄ±f klasÃ¶rÃ¼ bulunamadÄ±: {self.data_dir}")
        
        # TÃ¼m gÃ¶rÃ¼ntÃ¼leri topla
        all_images = []
        all_labels = []
        
        for cls_name in self.classes:
            cls_path = self.data_dir / cls_name
            image_files = list(cls_path.glob('*.jpg')) + list(cls_path.glob('*.jpeg')) + \
                         list(cls_path.glob('*.png')) + list(cls_path.glob('*.JPG')) + \
                         list(cls_path.glob('*.JPEG')) + list(cls_path.glob('*.PNG'))
            
            for img_file in image_files:
                all_images.append(str(img_file))
                all_labels.append(self.class_to_idx[cls_name])
        
        if len(all_images) == 0:
            raise ValueError(f"HiÃ§ gÃ¶rÃ¼ntÃ¼ bulunamadÄ±: {self.data_dir}")
        
        # Train/Val split
        indices = np.arange(len(all_images))
        np.random.seed(42)
        np.random.shuffle(indices)
        split_idx = int(len(indices) * train_ratio)
        
        if split == 'train':
            self.images = [all_images[i] for i in indices[:split_idx]]
            self.labels = [all_labels[i] for i in indices[:split_idx]]
        else:
            self.images = [all_images[i] for i in indices[split_idx:]]
            self.labels = [all_labels[i] for i in indices[split_idx:]]
        
        if len(self.images) == 0:
            raise ValueError(f"{split} dataset boÅŸ! Toplam gÃ¶rÃ¼ntÃ¼: {len(all_images)}, Split idx: {split_idx}")
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        img_path = self.images[idx]
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (IMAGE_SIZE, IMAGE_SIZE), color='black')
        
        label = self.labels[idx]
        
        if self.transform:
            image = self.transform(image)
        
        return image, label

# Data Augmentation ve Transform - HÄ±zlÄ± eÄŸitim iÃ§in optimize edilmiÅŸ
train_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(10),  # 15'ten 10'a dÃ¼ÅŸÃ¼rÃ¼ldÃ¼ (daha hÄ±zlÄ±)
    transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.15),  # Hafifletildi
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # ImageNet normalization
])

val_transform = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Dataset ve DataLoader oluÅŸtur
try:
    train_dataset = ImageDataset(train_path, transform=train_transform, split='train', train_ratio=TRAIN_RATIO)
    val_dataset = ImageDataset(train_path, transform=val_transform, split='val', train_ratio=TRAIN_RATIO)
    
    # BoÅŸ dataset kontrolÃ¼
    if len(train_dataset) == 0:
        raise ValueError(f"Train dataset boÅŸ! Veri seti yapÄ±sÄ±nÄ± kontrol edin: {train_path}")
    if len(val_dataset) == 0:
        raise ValueError(f"Validation dataset boÅŸ! Veri seti yapÄ±sÄ±nÄ± kontrol edin: {train_path}")
    
    print(f"âœ… Dataset'ler oluÅŸturuldu:")
    print(f"  ğŸ“Š Train: {len(train_dataset)} gÃ¶rÃ¼ntÃ¼")
    print(f"  ğŸ“Š Validation: {len(val_dataset)} gÃ¶rÃ¼ntÃ¼")
    
    # SÄ±nÄ±f isimlerini al
    class_names = train_dataset.classes
    
    num_workers = 0 if platform.system() == 'Windows' else 2
    
    # DataLoader oluÅŸtur
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, 
                              num_workers=num_workers, pin_memory=True if torch.cuda.is_available() else False)
    val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, 
                            num_workers=num_workers, pin_memory=True if torch.cuda.is_available() else False)
    
    print(f"âœ… PyTorch Dataset hazÄ±r!")
    print(f"  ğŸ“Š Train: {len(train_dataset)} gÃ¶rÃ¼ntÃ¼")
    print(f"  ğŸ“Š Validation: {len(val_dataset)} gÃ¶rÃ¼ntÃ¼")
    print(f"  ğŸ“Š SÄ±nÄ±f sayÄ±sÄ±: {num_classes}")
    print(f"  ğŸ“‹ SÄ±nÄ±flar: {', '.join(class_names[:10])}")
    if len(class_names) > 10:
        print(f"      ... ve {len(class_names) - 10} sÄ±nÄ±f daha")
    
except Exception as e:
    print(f"âŒ Hata: {e}")
    print(f"\nğŸ’¡ Veri seti yapÄ±sÄ± kontrol ediliyor...")
    print(f"ğŸ“ Train path: {train_path}")
    train_path_obj = Path(train_path)
    if train_path_obj.exists():
        print(f"ğŸ“‚ Ä°Ã§erik:")
        for item in list(train_path_obj.iterdir())[:10]:
            if item.is_dir():
                images = list(item.glob('*.jpg')) + list(item.glob('*.jpeg')) + \
                        list(item.glob('*.png'))
                print(f"  ğŸ“ {item.name}/ ({len(images)} gÃ¶rÃ¼ntÃ¼)")
                if len(images) == 0:
                    for sub_item in list(item.iterdir())[:5]:
                        if sub_item.is_dir():
                            sub_images = list(sub_item.glob('*.jpg')) + list(sub_item.glob('*.jpeg')) + \
                                        list(sub_item.glob('*.png'))
                            print(f"    ğŸ“ {sub_item.name}/ ({len(sub_images)} gÃ¶rÃ¼ntÃ¼)")
    raise
```

```python
# PyTorch CNN Modeli - Conv2d katmanlarÄ± ile optimize edilmiÅŸ - Accuracy %90+ iÃ§in
class PyTorchCNN(nn.Module):
    """
    Scikit-learn benzeri katman yapÄ±sÄ± ile PyTorch CNN modeli
    Conv2d katmanlarÄ± optimize edilmiÅŸ, Accuracy %90+ hedefleniyor
    """
    def __init__(self, num_classes):
        super(PyTorchCNN, self).__init__()
        
        # Conv2d katmanlarÄ± - stride=2 ile optimize edilmiÅŸ
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=2, padding=1)  # stride=2
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool1 = nn.MaxPool2d(2, 2)
        self.dropout1 = nn.Dropout2d(0.25)
        
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1)  # stride=2
        self.bn3 = nn.BatchNorm2d(128)
        self.conv4 = nn.Conv2d(128, 128, kernel_size=3, padding=1)
        self.bn4 = nn.BatchNorm2d(128)
        self.pool2 = nn.MaxPool2d(2, 2)
        self.dropout2 = nn.Dropout2d(0.25)
        
        self.conv5 = nn.Conv2d(128, 256, kernel_size=3, padding=1)
        self.bn5 = nn.BatchNorm2d(256)
        self.conv6 = nn.Conv2d(256, 256, kernel_size=3, padding=1)
        self.bn6 = nn.BatchNorm2d(256)
        self.pool3 = nn.MaxPool2d(2, 2)
        self.dropout3 = nn.Dropout2d(0.25)
        
        # Global Average Pooling - Daha hÄ±zlÄ± eÄŸitim iÃ§in 512 katmanÄ± kaldÄ±rÄ±ldÄ±
        self.global_avg_pool = nn.AdaptiveAvgPool2d(1)
        
        # Fully Connected Layers - HÄ±zlÄ± eÄŸitim iÃ§in optimize edilmiÅŸ
        self.fc1 = nn.Linear(256, 128)  # 512->256 yerine 256->128
        self.dropout5 = nn.Dropout(0.5)
        self.fc2 = nn.Linear(128, num_classes)  # Bir FC katmanÄ± kaldÄ±rÄ±ldÄ±
    
    def forward(self, x):
        # 160x160 -> 80x80 (stride=2)
        x = F.relu(self.bn1(self.conv1(x)))
        x = F.relu(self.bn2(self.conv2(x)))
        x = self.pool1(x)  # 80x80 -> 40x40
        x = self.dropout1(x)
        
        # 40x40 -> 20x20 (stride=2)
        x = F.relu(self.bn3(self.conv3(x)))
        x = F.relu(self.bn4(self.conv4(x)))
        x = self.pool2(x)  # 20x20 -> 10x10
        x = self.dropout2(x)
        
        # 10x10 -> 5x5
        x = F.relu(self.bn5(self.conv5(x)))
        x = F.relu(self.bn6(self.conv6(x)))
        x = self.pool3(x)  # 5x5 -> 2x2
        x = self.dropout3(x)
        
        # Global Average Pooling - Daha hÄ±zlÄ± eÄŸitim iÃ§in
        x = self.global_avg_pool(x)
        x = x.view(x.size(0), -1)
        
        # Fully Connected - Ä°ki katman (daha hÄ±zlÄ±)
        x = F.relu(self.fc1(x))
        x = self.dropout5(x)
        x = self.fc2(x)
        
        return x

# Model oluÅŸtur
model = PyTorchCNN(num_classes=num_classes).to(device)

# Model parametrelerini say
total_params = sum(p.numel() for p in model.parameters())
print(f"ğŸ“Š PyTorch Model Bilgileri:")
print(f"  Toplam parametre: {total_params:,}")
print(f"  SÄ±nÄ±f sayÄ±sÄ±: {num_classes}")

# Loss ve Optimizer - HÄ±zlÄ± eÄŸitim iÃ§in optimize edilmiÅŸ
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.002, weight_decay=1e-4)  # LR artÄ±rÄ±ldÄ± (daha hÄ±zlÄ± Ã¶ÄŸrenme)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=3, verbose=True)

print("âœ… PyTorch modeli hazÄ±r!")
```

<li><strong>Epoch</strong>: TÃ¼m eÄŸitim verisinin bir kez modelden geÃ§irilmesi</li>
<li><strong>Batch</strong>: Her iterasyonda iÅŸlenen gÃ¶rÃ¼ntÃ¼ grubu</li>
<li><strong>Early Stopping</strong>: Model iyileÅŸmediÄŸinde durdurur (zaman ve kaynak tasarrufu)</li>
<li><strong>Best Model</strong>: En yÃ¼ksek validation accuracy'ye sahip model kaydedilir</li>

```python
# PyTorch Model EÄŸitimi - Accuracy %90+ hedefleniyor
print("ğŸš€ PyTorch Model EÄŸitimi BaÅŸlÄ±yor...\n")
print("ğŸ¯ Hedef: Loss dÃ¼ÅŸÃ¼k, Accuracy %90+\n")

def train_epoch(model, loader, criterion, optimizer, device, scaler=None):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0
    
    for images, labels in tqdm(loader, desc='Training'):
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        
        # Mixed Precision Training
        if scaler is not None:
            with autocast():
                outputs = model(images)
                loss = criterion(outputs, labels)
            
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
        else:
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
        
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / len(loader)
    epoch_acc = 100 * correct / total
    return epoch_loss, epoch_acc

def validate(model, loader, criterion, device):
    model.eval()
    running_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in tqdm(loader, desc='Validation'):
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    epoch_loss = running_loss / len(loader)
    epoch_acc = 100 * correct / total
    return epoch_loss, epoch_acc

# EÄŸitim dÃ¶ngÃ¼sÃ¼
train_losses = []
train_accs = []
val_losses = []
val_accs = []

num_epochs = 25  # HÄ±zlÄ± eÄŸitim iÃ§in optimize edilmiÅŸ (30'dan 25'e dÃ¼ÅŸÃ¼rÃ¼ldÃ¼)
best_val_acc = 0.0
best_val_loss = float('inf')
patience = 7
patience_counter = 0

print(f"ğŸ“Š EÄŸitim bilgileri:")
print(f"  - Train Ã¶rnekleri: {len(train_dataset)}")
print(f"  - Validation Ã¶rnekleri: {len(val_dataset)}")
print(f"  - SÄ±nÄ±f sayÄ±sÄ±: {num_classes}")
print(f"  - Batch size: {BATCH_SIZE}")
print(f"  - GÃ¶rÃ¼ntÃ¼ boyutu: {IMAGE_SIZE}x{IMAGE_SIZE}")
print(f"  - Toplam epoch: {num_epochs}")
print(f"  - Early Stopping patience: {patience}\n")

for epoch in range(num_epochs):
    print(f"\n{'='*70}")
    print(f"Epoch {epoch+1}/{num_epochs}")
    print(f"{'='*70}")
    
    train_loss, train_acc = train_epoch(model, train_loader, criterion, optimizer, device, scaler)
    val_loss, val_acc = validate(model, val_loader, criterion, device)
    
    scheduler.step(val_loss)
    
    train_losses.append(train_loss)
    train_accs.append(train_acc)
    val_losses.append(val_loss)
    val_accs.append(val_acc)
    
    print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}%")
    print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
    
    # En iyi modeli kaydet
    if val_acc > best_val_acc or (val_acc == best_val_acc and val_loss < best_val_loss):
        best_val_acc = val_acc
        best_val_loss = val_loss
        torch.save(model.state_dict(), 'best_model.pth')
        print(f"âœ… Yeni en iyi model kaydedildi! (Acc: {best_val_acc:.2f}%, Loss: {best_val_loss:.4f})")
        patience_counter = 0
    else:
        patience_counter += 1
    
    # Early Stopping
    if patience_counter >= patience:
        print(f"\nâ¹ï¸ Early Stopping: {patience} epoch boyunca iyileÅŸme olmadÄ±")
        break
    
    # Accuracy %90+ hedefine ulaÅŸÄ±ldÄ± mÄ±?
    if val_acc >= 90.0:
        print(f"\nğŸ¯ Hedef baÅŸarÄ±ldÄ±! Accuracy: {val_acc:.2f}% >= 90%")
        if val_loss < 0.5:
            print(f"âœ… Loss deÄŸeri de dÃ¼ÅŸÃ¼k: {val_loss:.4f}")
            print("ğŸ’¡ Ä°stediÄŸiniz hedeflere ulaÅŸÄ±ldÄ±!")
        break

print(f"\nğŸ‰ EÄŸitim tamamlandÄ±!")
print(f"ğŸ† En iyi Validation Accuracy: {best_val_acc:.2f}%")
print(f"ğŸ† En iyi Validation Loss: {best_val_loss:.4f}")

# En iyi modeli yÃ¼kle
if os.path.exists('best_model.pth'):
    model.load_state_dict(torch.load('best_model.pth'))
    print("âœ… En iyi model yÃ¼klendi.")
```
